## Algorytm propagacji wstecznej błędu

### Oznaczenia

- Omówmy sieć $M$-warstwową, $m=1,2,\ldots,M$ i używamy $V_i^m$ jako wyjścia jednostki $i$-tej w warstwie $m$-tej
- $V_j^0$ będzie synonimem $\xi_i$, wejścia $i$-tego
- Zwracamy uwagę, że górny indeks oznacza $m$-tą warstwę, a nie wzorzec
- Wielkość $w_{ij}^{m}$ jest połączeniem od $V_j^{m-1}$ do $V_j^m$

### Kroki algorytmu

- Początkowe wartości wag wybierz jako małe liczby losowe
- Wybierz wzorzec $\xi_{k}^{\mu}$ i podaj go na wejście ( warstwa
      $m=0$) tak , aby $$V_{k}^{0} = \xi_{k}^{\mu} \text { dla każdego
      $k$}$$
- Przepuść sygnał przez sieć do przodu za pomocą
      $$V_{i}^{m} = g(h_i^m) = g(\sum_{j} w_{ij}^m V_{j}^{m-1})$$
      dla każdego $i$ oraz $m$ , dopóki nie obliczysz wszystkich
      wyjść $V_i^M$.
- Oblicz błędy $\delta$ w warstwie wyjściowej
      $$\delta_i^M = g'(h_i^M)\left[ \zeta_i^\mu - V_i^M \right]$$
      porównując wyjścia otrzymane $V_i^M$ z pożądanym
      $\zeta_i^\mu$ dla wzorca $\mu$ -tego, który został podany.
- Oblicz błędy $\delta$ w poprzednich warstwach za pomocą
      propagacji wstecznej błędu
      $$\delta_{i}^{m-1} = g'(h_{i}^{m-1})\sum_j w_{ji}^{m}\delta_j^m$$
      dla $m=M,M-1,\ldots,2$ , dopóki nie obliczysz błędów dla
      wszystkich jednostek.
- Zastosuj $$\Delta w_{ij}^m = \eta \delta_i^m V_j^{m-1}$$
      do modyfikacji wszystkich połączeń według zależności
      $w_{ij}^{nowe} = w_{ij}^{stare} + \Delta w_{ij}$.
- Wróć do punktu 2 i powtórz procedurę dla następnego
      wzorca.

````python
import numpy as np

# seed random numbers to make calculation deterministic
np.random.seed(42069)

def sigmoid(x):
    return 1/(1+np.exp(-x))

def sigmoid_derivative(x):
    return x*(1-x)

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2    # 1 - x**2

def mean_squared_error(y_true, y_pred): 
    return np.mean((y_true - y_pred) ** 2)

class Layer:
    def __init__(self, units, activation, input_dim):
        self.units = units
        self.activation = activation
        self.input_dim = input_dim
        # random weights from -1 to 1
        self.weights = np.random.rand(input_dim, units).T * 2 - 1
        self.bias = np.zeros((units, 1))
        self._input = None
        self._output = None

    def __str__(self):
        return f"Layer: {self.units} units, {self.activation}\n"

    def __repr__(self):
        return f"Layer: {self.units} units, {self.activation}\n"

    def feedforward(self, input):
        self._input = input
        if self.activation == "sigmoid":
            self._output = sigmoid(np.dot(self.weights, self._input) + self.bias)
        elif self.activation == "relu":
            self._output = relu(np.dot(self.weights, self._input) + self.bias)
        elif self.activation == "tanh": 
            self._output = tanh(np.dot(self.weights, self._input) + self.bias)
        else:
            raise ValueError("Invalid activation function")

        return self._output

    def backprop(self, delta, weights, eta):
        if self.activation == "sigmoid":
            return self._sigmoid_backprop(delta, weights, eta)
        elif self.activation == "relu":
            return self._relu_backprop(delta, weights, eta)
        elif self.activation == "tanh":
            return self._tanh_backprop(delta, weights, eta)
        else:
            raise ValueError("Invalid activation function")

    def _sigmoid_backprop(self, delta, weights, eta):
        delta_new = np.dot(weights.T, delta) * sigmoid_derivative(self._output)
        self.update_weights(delta_new, eta)

        return delta_new

    def _relu_backprop(self, delta, weights, eta):
        delta_new = np.dot(weights.T, delta) * relu_derivative(self._output)
        self.update_weights(delta_new, eta)

        return delta_new
    
    def _tanh_backprop(self, delta, weights, eta):
        delta_new = np.dot(weights.T, delta) * tanh_derivative(self._output)
        self.update_weights(delta_new, eta)

        return delta_new

    def update_weights(self, delta, eta):
        self.weights += eta * np.dot(delta, self._input.T)
        self.bias += eta * np.sum(delta, axis=1, keepdims=True)/self.units


class NeuralNetwork:
    def __init__(self):
        self.layers = []
        self.eta = 0.1

    def _feedforward(self, input):
        # feed the input through the network
        output = input

        for layer in self.layers:
            output = layer.feedforward(output)

        return output

    def add(self, units, activation, input_dim):
        # add a layer to the network

        self.layers.append(Layer(units, activation, input_dim))

    def summary(self):
        # return a summary of the network as a string
        summary = "Neural Network Summary\n"
        summary += f"Learning Rate: {self.eta}\n"
        summary += "Layers:\n"

        for (i, layer) in enumerate(self.layers):
            summary += f"Layer {i}: {layer.units} units, {layer.activation}\n"
            
        return summary

    def predict(self, x):
        return self._feedforward(x.T)

    def set_learning_rate(self, eta):
        self.eta = eta

    def fit(self, x, y, epochs):
        input = x.T

        for _ in range(epochs):
            output = self._feedforward(input)
            self._backprop(input, output, y.T)

    def _backprop(self, input, output, y):
        # handle last layer separately
        if self.layers[-1].activation == "sigmoid":
            delta = (y - output) * sigmoid_derivative(output)
        elif self.layers[-1].activation == "relu":
            delta = (y - output) * relu_derivative(output)
        elif self.layers[-1].activation == "tanh":  
            delta = (y - output) * tanh_derivative(output)
        else:
            raise ValueError("Invalid activation function")
        
        weights = self.layers[-1].weights
        self.layers[-1].update_weights(delta, self.eta)

        for i in range(len(self.layers)-2, -1, -1):
            delta = self.layers[i].backprop(delta, weights, self.eta)
            weights = self.layers[i].weights
````

````python
import numpy as np
import matplotlib.pyplot as plt
from concurrent.futures import ProcessPoolExecutor
import os

# change seed
np.random.seed(42069)

# Create datasets
datasets = {
    "XOR": (np.array([[0, 0], [0, 1], [1, 0], [1, 1]]), np.array([[0], [1], [1], [0]])),
    "AND": (np.array([[0, 0], [0, 1], [1, 0], [1, 1]]), np.array([[0], [0], [0], [1]])),
    "OR": (np.array([[0, 0], [0, 1], [1, 0], [1, 1]]), np.array([[0], [1], [1], [1]])),
}

# Parameters to test
learning_rates = [0.001, 0.01, 0.1, 0.5]
epoch_sizes = [1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000]
architectures = {
    "2-8-1": [(8, "sigmoid", 2), (1, "sigmoid", 8)],
    "2-8-1_relu": [(8, "relu", 2), (1, "relu", 8)],
    "2-8-8-1": [(8, "sigmoid", 2), (8, "sigmoid", 8), (1, "sigmoid", 8)],
    "2-8-8-1_mixed": [(8, "relu", 2), (8, "tanh", 8), (1, "sigmoid", 8)],
}

results = {}


def train_network(params):
    function, arch_name, lr, epochs, arch = params
    x, y = datasets[function]

    # Initialize network
    nn = NeuralNetwork()
    nn.set_learning_rate(lr)
    for units, activation, input_dim in arch:
        nn.add(units, activation, input_dim)

    # Train network
    nn.fit(x, y, epochs)
    pred = nn.predict(x)
    mse = mean_squared_error(y.T, pred)

    return (function, arch_name, lr, epochs, mse)

# Use ProcessPoolExecutor to handle multiprocessing
if __name__ == "__main__" or os.environ.get("JUPYTERHUB_USER") is not None:
    executor = ProcessPoolExecutor(
        max_workers=4
    )  # Adjust the number of workers as needed

    tasks = []
    for function, data in datasets.items():
        for arch_name, arch in architectures.items():
            for lr in learning_rates:
                for epochs in epoch_sizes:
                    tasks.append((function, arch_name, lr, epochs, arch))

    # Submit tasks to the executor
    futures = [executor.submit(train_network, task) for task in tasks]

    # Collect results as they complete
    results = [future.result() for future in futures]
    executor.shutdown()

    # Organize results for plotting
    organized_results = {}
    for function, arch_name, lr, epochs, mse in results:
        if function not in organized_results:
            organized_results[function] = {}
        if arch_name not in organized_results[function]:
            organized_results[function][arch_name] = {}
        organized_results[function][arch_name][(lr, epochs)] = mse
````

````python
# Plotting results
fig, axs = plt.subplots(
        len(datasets), len(architectures), figsize=(15, 10), sharey=True
    )
fig.suptitle("MSE by Learning Rate and Epoch for Various Functions and Architectures")

lines = []  # To hold the line objects for the legend
labels = []  # To hold the label names for the legend

for i, function in enumerate(datasets):
    for j, arch in enumerate(architectures):
        ax = axs[i][j]
        data = organized_results[function][arch]
        for idx, lr in enumerate(learning_rates):
            mserrors = [data[(lr, epochs)] for epochs in epoch_sizes]
            (line,) = ax.plot(epoch_sizes, mserrors, label=f"LR={lr}", marker=".")
            if i == 0 and j == 0:  # Only add to legend once
                lines.append(line)
                labels.append(f"LR={lr}")
        ax.set_title(f"{function} - {arch}")
        ax.set_xlabel("Epochs")
        ax.set_ylabel("MSE")

# Placing a single legend outside the last subplot
fig.legend(
    lines, labels, loc="upper right", bbox_to_anchor=(1.1, 1), title="Learning Rates"
)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
````
